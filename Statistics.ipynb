```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
import statistics

x = np.array([19.690, 20.079, 19.401, 20.162, 19.776])
y = np.array([20.555, 20.551,20.796, 21.152, 20.088 ])

# a) la moyen

print('la moyen')
meanX = statistics.mean(x)
meanY = statistics.mean(y)

print('moyen de x =',meanX)

print(meanY)


#la variance
print('la variance')
varX = np.var(x)
varY = np.var(y)
print('variance de x',varX)
print( varY)

#calcule de l'ecart type
print(' ')
print('l\'ecart type')


print('ecart type de x = ')
ecart_typeX = statistics.pstdev(x)

print(ecart_typeX)

print('ecart type de y')
ecart_typeY = statistics.pstdev(y)
print(ecart_typeY)

# b) Calculer le pourcentage de différence entre les deux moyennes
print('')
print('pourcentage de différence entre les deux moyennes')
x_bar = sum(x)/5
y_bar = sum(y)/5

print('x_bar')
print(x_bar)

pourcentage_difference = ((x_bar - y_bar ) / x_bar ) * 100
print('pourcentage_difference = ')
print(pourcentage_difference)

print('Commentaire sur le pourcentage de différence entre les deux moyennes')
print( 'les hommes (y) touches  en moyen un salaire > aux femmes (x) de 4%')

print("2) Modélisation ")
# a) Afficher comme un nuage de points les données ci-dessus, on mettra en X le salaire des femmes 
#et en Y celui des hommes.
print(" a) le Nuage de points : ")
plt.title('nuage de points')
plt.scatter(x, y, alpha=1)
plt.show()

```

```python

#b) Calculer les coefficients de la droite des moindres-carrés correspondant à ces données.
# y = alphax + b
print(" b) ")
print("l'équation de la ligne droite : y = alphax + b, donc on va calculer les coefficients apha et b")
# alpha = covarianc / variance et b = alpha*x
print("calculons d'abord la covariance")
covariance = np.sum(x*y)- 5*meanX*meanY
print('la covariance = ' , covariance)

alpha = covariance / varX
print('le cofficient apha = ', alpha)

b = meanY - alpha*meanX
print("le cofficient b = ", b)

#  donc la droite Y = 1.61 x + (-11.35)
Y = alpha*x + b
print("la droite Y = 1.61 x - 11.35 , donc en fonction de la valeur de x on obtien : ",Y)
# c) Représenter cette droite sur le nuage de points.
print("c) Représentation de la droite sur le nuage de points. ")

plt.scatter(x, y, color = 'blue')
plt.plot(x,Y, color = 'green')
plt.xlabel('Les femmes')
plt.ylabel('les hommes')
plt.show()

#  Commenter la valeur du coefficient de corrélation et des coefficients de la droite. Que dire de la
# méthode ?
print("Commenter la valeur du coefficient de corrélation et des coefficients de la droite. ")
print("on va calculer d'abord le coefficient de corellation")
corellation = covariance / ecart_typeX*ecart_typeY
print("corellation = ", corellation)

print("On peut constater que le coefficient de corellation est proche de -1, \n donc on peut dire que la relation lineaire entre les deux variable est pas forte. Enfin on peut conclure que la méthode de moindre carré n'est pas adapté à cette situation")
```
